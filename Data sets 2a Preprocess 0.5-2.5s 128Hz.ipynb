```python
import mne
import scipy
import numpy as np
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import matplotlib
import keras

def read_raw(path):
    raw = mne.io.read_raw_gdf(path,preload=True,eog=['EOG-left', 'EOG-central', 'EOG-right'])
    raw.drop_channels(['EOG-left', 'EOG-central', 'EOG-right'])
    raw.resample(128)
    raw.set_eeg_reference()
    events = mne.events_from_annotations(raw)
    return raw,events
    
def preprocess():
    labels_list=[]
    for i in range(1,10):
        raw,events = read_raw(f'BCICIV_2a_gdf/A0{i}T.gdf')
        lables = []
        largest_id = np.max(events[0][:, 2])
        event_id = [largest_id - 3, largest_id - 2, largest_id - 1, largest_id]
        epochs = mne.Epochs(raw, events[0], event_id, tmin = 0.5, tmax = 2.5 - 1 / raw.info['sfreq'], baseline=(0.5, 0.7), on_missing ='warn')
        lables = epochs.events[:,-1]
        lables = np.subtract(lables, largest_id - 3)
        data = epochs.get_data()
        scaler = mne.decoding.Scaler(scalings='mean')
        data_scaled = scaler.fit_transform(data)
        for j in range(len(data)):
            np.save(f"Data sets 2a 0.5-2.5s 128Hz/data/A{str(i).rjust(2, '0')}TE{str(j).rjust(4, '0')}.npy", data_scaled[j])
        labels_list.append(lables)
    for i in range(1,10):
        raw,events = read_raw(f'BCICIV_2a_gdf/A0{i}E.gdf')
        matdata = scipy.io.loadmat(f'BCICIV_2a_gdf/true_labels/A0{i}E.mat')
        lables = np.subtract(matdata['classlabel'], 1).reshape(-1)
        largest_id = np.max(events[0][:, 2])
        event_id = [largest_id]
        epochs = mne.Epochs(raw, events[0], event_id, tmin = 0.5, tmax = 2.5 - 1 / raw.info['sfreq'], baseline=(0.5, 0.7), on_missing ='warn')
        data = epochs.get_data()
        scaler = mne.decoding.Scaler(scalings='mean')
        data_scaled = scaler.fit_transform(data)
        for j in range(len(data)):
            np.save(f"Data sets 2a 0.5-2.5s 128Hz/data/A{str(i).rjust(2, '0')}EE{str(j).rjust(4, '0')}.npy", data_scaled[j])
        labels_list.append(lables)
    labels_list = np.concatenate(labels_list)
    labels_list = keras.utils.to_categorical(labels_list, 4)
    np.save('Data sets 2a 0.5-2.5s 128Hz/labels_list.npy', labels_list)
preprocess()
labels_list = np.load('Data sets 2a 0.5-2.5s 128Hz/labels_list.npy')
print(labels_list)
print(len(labels_list))
```
